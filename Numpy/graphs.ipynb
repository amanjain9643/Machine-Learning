```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
x=np.linspace(-10,10,100)
y=x
```

```python
plt.plot(x,y)
```

```python
x=np.linspace(-10,10,100)
y=x**2
plt.plot(x,y)
```

```python
x=np.linspace(-10,10,100)
y=np.sin(x)
plt.plot(x,y)
```

```python
x=np.linspace(-10,10,100)
y=np.cos(x)
plt.plot(x,y)
```

```python
x=np.linspace(-10,10,100)
y=x*np.log(x)
plt.plot(x,y)
```

```python
x=np.linspace(-10,10,100)
y=np.exp(x)
plt.plot(x,y)
```

```python
def sigmoid(array):
    return 1/(1+(np.exp(array))**-1)
```

```python
x=np.linspace(-10,10,100)
y=sigmoid(x)
plt.plot(x,y)
```
